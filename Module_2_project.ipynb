# Explanation of the differents enhancement.

- Noise reduction


A median filter is applied to each R,G,B channel in order to remove noise. A median filter is simply a filter that takes one pixel and replace its value by the median value of its neighbors pixels, you can define the number of neighbors. This way, you can remove outliers. We apply this to every pixel in the image.
The number of neighbors pixels can be set by increasing or decreasing a parameter in the program, so depending on the noise you can adapt the filter.

- Auto-level

A auto-level transformation is applied to the image. The user can set a Tmin and Tmax threshold. Every pixel value before Tmin is set to 0 and every pixel value after Tmax is set to 255. Between Tmin and Tmax, we create a straight line that starts in Tmin at the value 0 and end in Tmax at the value 255. 
So every low value pixel (< Tmin) will appears dark, every high value pixel (>Tmax) will appear white and the pixel between them will follow the straight line. This way you can improve contrat.

In this project, as we are using RGB image, we will convert RGB to HSV and make an auto-level on the value channel of HSV. Indeed, auto-level is related to the overall luminosity on the image so it seems more logical to work on the value channel which is basically the gray level image.

- Equalization of the value

A equalization transformation is a transformation that makes the histogram of the image as flat as possible. The transformation function is the cumulative histogram. This way, if 80% of the pixels have a value below 50, it will project thoses pixels between 0 and 204 (80%*255). It will improve contrast since the difference between two pixel value will increase. By example a pixel that has a value of 30 and a pixel that has a value of 40 might be changed to 85 and 150, increasing the differences between the two pixels and therefore the contrast. 

We will convert RGB to HSV and apply equalization on the value channel of HSV since equalization is generally related to the overall luminosity on the image.

- Equalization of the saturation

We will make exactly the same as we did above, but we will apply the equalization on the saturation channel of HSV. It is a bit atypical but it gives real good visual effects, like this :

- Equalization of the R,G,B channel

Same as above, we just apply the equalization to each R,G,B channel and it gives a nice color effect.

- Gamma correction (automatically or manually set)

The Gamma correction allows to change the contrast for the low or the high level depending on the gamma value. 
If gamma is < 1, the low level are contrasted and if gamma is > 1 the high level are contrasted. So this correction allows you improve contrast in low or high level.  

The gamma value can be automatically found. For that it converts the image into HSV and takes the value channel. We use the value channel because it is basically  the gray scale image and gives a really good idea of the brightness of the image. It then compare the mid value of the gray scale (255*0.5) with the mean value of the value channel to give the best gamma value. 

- Color temperature

When a object emits lights, depending on the temperature of the object, the wavelenght of the light emitted will change, so the color that you see will be different. This is translated into a R,G,B triplet depending on the temperature. 
There are tables that makes this conversion as the relation between color and temperature isn't linear. I found somebody who converted the temperature from 0K to 10000K by step or 500K. 
Link : https://stackoverflow.com/questions/11884544/setting-color-temperature-for-a-given-image-like-in-photoshop
So you can simply convert a temperature into a R,G,B triplet and then apply a linear transformation to every pixel on the image by multiplying each pixel with the triplet then normalize it. 

- Paint looking picture (make your picture look like it was painted by someone)

During the labs, I noticed that applying a closing transformation to each R,G,B channel gives an effect like the image was painted by somebody. So the user can activate this option and choose the radius of the filter to increase or decrease this effect.
A closing is a dilation followed by an erosion. First it dilates the pixel and then it erodes it. The strengh of the closing depends on the structuring element size which can be modified by the user.

```python
#LIBRARY 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from skimage.io import imread,imshow,imsave
from skimage.filters.rank import median,mean
from skimage.morphology import disk, square, opening, closing, erosion, dilation, label
from skimage.color import rgb2hsv, hsv2rgb
import cv2
import math
from PIL import Image
import random 


#BASIC FUNCTION 




def turn2D_into_RGB(im):
    return np.dstack((im, im, im))

def save_im(im, filepath):
    imsave(filepath, im)    
    
def display_im(im,title):
    plt.figure(figsize=(10, 10))
    plt.imshow(im)
    plt.title(title)
    plt.show()

def rescale_image_to_255(im):
    
    
    if (im.ndim ==3):
        im_R = im[:,:,0]/np.max(im[:,:,0])
        im_G = im[:,:,1]/np.max(im[:,:,1])
        im_B = im[:,:,2]/np.max(im[:,:,2])

        return (np.dstack((im_R, im_G, im_B))*255).astype("uint8")
    else :
        im = im/np.max(im)
        return (im*255).astype("uint8")
    
def normal_hist(im):
    hist, bins = np.histogram(im.flatten(), bins = range(257))
    return hist/np.sum(hist), bins
def cumult_hist(hist):
    cumul_hist = np.zeros(256)
    for x in range (0,256,1):
        cumul_hist[x] = np.sum(hist[0:x])
    return cumul_hist
def display_hist(im):
    hist, bins = normal_hist(im)
    plt.figure(figsize=(10, 10))
    plt.bar(bins[:-1], hist)
    plt.show()
    
    

def add_noise(im): #To try the noise cancelling filter

    for y in range(im.shape[0]):
        for x in range(im.shape[1]):
            if (random.random() < 0.1):  
                    im[y,x] =im[y,x] * random.random()
    return im



###EQUALIZATION

    
    
def equalize_channel(im):
    resc_im = rescale_image_to_255(im)
    hist, bins = normal_hist(resc_im)
    cumul_hist = cumult_hist(hist)
    lut = (cumul_hist*255).astype('uint8')
    return lut[resc_im]



def equalize_RGB(im):
    eq_R = equalize_channel(im[:,:,0])
    eq_G = equalize_channel(im[:,:,1])
    eq_B = equalize_channel(im[:,:,2])
    return np.dstack((eq_R, eq_G, eq_B))



def equalize_value_image(im):
    hsv_img = rgb2hsv(im)
    value_img= (hsv_img[:, :, 2])
    new_hsv = np.dstack(((hsv_img[:, :, 0]*255).astype('uint8'),(hsv_img[:, :, 1]*255).astype('uint8'),
                     equalize_channel(value_img)))
    eq_im = hsv2rgb(new_hsv)
    return eq_im    


def equalize_saturation_image(im):
    hsv_img = rgb2hsv(im)
    sat_im= (hsv_img[:, :, 1]*255).astype('uint8')
    new_hsv = np.dstack(((hsv_img[:, :, 0]*255).astype('uint8'), equalize_channel(sat_im),
                    (hsv_img[:, :,2]*255).astype('uint8')))
    eq_im = hsv2rgb(new_hsv)
    return eq_im




###AUTO_LEVEL



def auto_level_channel(im, Tmin, Tmax):
    im = rescale_image_to_255(im)
    lut = np.zeros(256)
    lut [:Tmin] =0
    lut[Tmax:256] = 255
    lut_t = lut
    pente = (255-0)/(Tmax-Tmin) 
    for x in range(Tmin+1, Tmax,1):
        lut[x] = int(pente * (x-Tmin))
    return lut[im].astype('uint8')



def auto_level_value_image(im, Tmin, Tmax):
    hsv_img = rgb2hsv(im)
    value_img= (hsv_img[:, :, 2]*255).astype('uint8')
    auto_level = auto_level_channel(value_img, Tmin, Tmax)
    new_hsv = np.dstack(((hsv_img[:, :, 0]*255).astype('uint8'),(hsv_img[:, :, 1]*255).astype('uint8'),
                     auto_level))
    return hsv2rgb(new_hsv)



###GAMMA



def find_best_gamma(img):
    
    hsv_img = rgb2hsv(img)
    val = (hsv_img[:,:,2]*255).astype('uint8')
    mid = 0.5*255
    mean = np.mean(val)
    gamma = math.log(mid)/math.log(mean)
    print(gamma)
    
    return gamma

def gamma_correction(im, gamma):
    #gamma correction is about luminosity so use value channel
    invGamma = 1.0/gamma
    lut = np.arange(256)
    lut = (((lut / 255.0) ** invGamma)*255).astype("uint8")   
    return lut[rescale_image_to_255(im)]



###PAINT



def give_paint_effect(im, radius):
   


    disk_im = disk(radius)

    im_R = im[:,:,0]
    im_G = im[:,:,1]
    im_B = im[:,:,2]
    
    op_R = closing(im_R, disk_im)
    op_G = closing(im_G, disk_im)
    op_B = closing(im_B, disk_im)


    return np.dstack((op_R, op_G, op_B))

   



#NOISE 



def remove_noise(im, radius):
    

    im_R = median(im[:,:,0], disk(radius))
    im_G = median(im[:,:,1], disk(radius))
    im_B = median(im[:,:,2], disk(radius))

    return np.dstack((im_R, im_G, im_B))



###TEMPERATURE



def change_temperature(im, Temp):
    r, g, b = kelvin_table[Temp]

    im_ = rescale_image_to_255(im)
    r = (im_[:,:,0] * (r/255.0))
    g =(im_[:,:,1] * (g/255.0))
    b= (im_[:,:,2] * (b/255.0))
    return (np.dstack((r, g, b))).astype("uint8")

kelvin_table = {
    1000: (255,56,0),
    1500: (255,109,0),
    2000: (255,137,18),
    2500: (255,161,72),
    3000: (255,180,107),
    3500: (255,196,137),
    4000: (255,209,163),
    4500: (255,219,186),
    5000: (255,228,206),
    5500: (255,236,224),
    6000: (255,243,239),
    6500: (255,249,253),
    7000: (245,243,255),
    7500: (235,238,255),
    8000: (227,233,255),
    8500: (220,229,255),
    9000: (214,225,255),
    9500: (208,222,255),
    10000: (204,219,255)}



    
```
